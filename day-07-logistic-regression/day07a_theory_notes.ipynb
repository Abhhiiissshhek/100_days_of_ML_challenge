# **WHY LINEAR REGRESSION FAILS**

* Linear Regression predicts ANY number.
* But we need PROBABILITY between 0 and 1

## LINEAR = ❌ FOR CLASSIFICATION

# **THE SIGMOID SOLUTION**

### **Sigmoid: 1 / (1 + e^-z)**

* z = 0   → 0.5
* z = 5   → 0.99
* z = -5  → 0.01

### SIGMOID takes ANY number and squishes it between 0 and 1.
* **This is the MAGIC!**

## **SIGMOID = ∞ → 0-1 📈**

# **THE FORMULA**

z = (b0) + (b1 * x1) + (b2 * x2) + (b3 * x3)

Probability = 1 / (1 + e^-z)

* First, LINEAR COMBINATION: multiply features by weights.
* Then pass through SIGMOID to get PROBABILITY

# **LOSS FUNCTION (How wrong are we?)**

* True=1, Predict=0.9 → Loss small ✅
* True=1, Predict=0.1 → Loss HUGE ❌
* True=0, Predict=0.1 → Loss small ✅
* True=0, Predict=0.9 → Loss HUGE ❌

* LOSS measures how WRONG we are.
* Big loss = bad. Small loss = good.
* Model tries to MINIMIZE loss

## **LOSS = HOW WRONG WE ARE 📉**

# **GRADIENT DESCENT (Learning)**

* GRADIENT DESCENT = rolling downhill.
* Start at top (high loss), take steps DOWN
* until you reach bottom (low loss)

## **GRADIENT DESCENT = LEARNING ⬇️**

# **WHAT WEIGHTS MEAN**


Feature	- Weight	-  Meaning
* Sex (female)	-> +2.5	-> ⬆️ Survive
* Alone	-> -1.5	-> ⬇️ Die
* Age	-> -0.4	-> ⬇️ Slightly

* Positive weight = increases survival.
* Negative weight = decreases survival.
* Bigger number = stronger impact

## **WEIGHTS TELL THE STORY 📊**
