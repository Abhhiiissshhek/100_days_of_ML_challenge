```python
import pandas

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pandas.read_csv(url)
print(df.head())
```

**"STEP 1: FIND MISSING VALUES üîç"**

```python
df.isnull().sum()
```

**177 | 687 | 2 MISSING**

Now let us calulate the damage % (percentage)

```python
(df.isnull().sum() / len(df)) * 100
```

**THE DROP DECISION (Cabin)**

">50% MISSING = DROP it üóëÔ∏è"

```python
# Cabin is 77% missing - basically useless
df.drop('Cabin', axis=1, inplace=True)
print("Cabin column dropped ‚ò†Ô∏è")
```

we dropped the cabin column

# **THE FILL DECISION (Age & Embarked)**

```python
# Age is 20% missing - fill with MEDIAN
df['Age'].fillna(df['Age'].median(), inplace=True)

# Embarked is 0.2% missing - fill with MODE
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

### **FILL: MEDIAN for numbers, MODE for categories ‚úÖ**

## THE SATISFYING FINISH (Cleaned Data)

```python
df.isnull().sum().sum()
```

**"0 MISSING VALUES üéâ"**

Our data is clean - 0 missing values. But raw columns aren't enough.

# **FEATURE 1: FAMILY SIZE**

```python
# Create Family Size from SibSp + Parch
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
print(df.head())
```

*  Passenger 1: FamilySize = 2  (traveling with spouse)
*  Passenger 6: FamilySize = 1  (alone)



**ALONE vs WITH FAMILY = different survival rates!**

# **Feature 2 : TITLE EXTRACTION**

```python
# Extract Title from Name
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```


1. Mr     =  517
2. Miss    = 182
3. Mrs     = 125
4. Master  =  40
5. Rare    =  27



TITLE = SOCIAL STATUS = KEY FEATURE!

# **FEATURE 3: AGE GROUPS**

```python
# Create Age Groups
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child', 'Teen', 'Adult', 'Middle-Aged', 'Senior']
df['AgeGroup'] = pandas.cut(df['Age'], bins=bins, labels=labels)
```

**BINNING continuous data = better patterns üìä**

# **FEATURE 4: IS_ALONE**

```python
# Create IsAlone flag
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
```

* IsAlone = 1 ‚Üí Passenger alone
* IsAlone = 0 ‚Üí With family

**SIMPLE FEATURES = POWERFUL INSIGHTS!**
