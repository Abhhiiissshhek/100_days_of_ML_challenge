# **SOLUTION: K-FOLD CROSS VALIDATION**
Data divided into 5 folds

* Fold 1: [TEST]+[TRAIN]+[TRAIN]+[TRAIN]+[TRAIN] → Score 82%
* Fold 2: [TRAIN]+[TEST]+[TRAIN]+[TRAIN]+[TRAIN] → Score 79%
* Fold 3: [TRAIN]+[TRAIN]+[TEST]+[TRAIN]+[TRAIN] → Score 85%
* Fold 4: [TRAIN]+[TRAIN]+[TRAIN]+[TEST]+[TRAIN] → Score 81%
* Fold 5: [TRAIN]+[TRAIN]+[TRAIN]+[TRAIN]+[TEST] → Score 83%

## **Average: 82% ± 2%**

┌─────┬─────┬─────┬─────┬─────┐
│  1  │  2  │  3  │  4  │  5  │
└─────┴─────┴─────┴─────┴─────┘

* Iteration 1: [███][░░░][░░░][░░░][░░░] → Test Fold 1
* Iteration 2: [░░░][███][░░░][░░░][░░░] → Test Fold 2
* Iteration 3: [░░░][░░░][███][░░░][░░░] → Test Fold 3
* Iteration 4: [░░░][░░░][░░░][███][░░░] → Test Fold 4
* Iteration 5: [░░░][░░░][░░░][░░░][███] → Test Fold 5

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

# Create dataset
X, y = make_classification(n_samples=1000, random_state=42)

model = LogisticRegression()

# 5-Fold Cross Validation
scores = cross_val_score(model, X, y, cv=5)

print("Scores for each fold:", scores)
print("Average score:", scores.mean())
```
